```python
import numpy as np
import pandas as pd
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('E:\DATA SCI AND ML\BirthAndPopulation.csv')
df1.head(10)
```

```python
df1['Population_change'] = df1.apply(lambda row: row.Births-row.Deaths, axis = 1)
df1.head()
```

```python
df1['birthRate'] = df1.apply(lambda row: row.Births/row.Population, axis = 1)
df1.head()
```

```python
df1['deathRate'] = df1.apply(lambda row: row.Deaths/row.Population, axis = 1)
df1.head()
```

```python
df1.head()
```

```python
df1['Population in 5 yrs'] = df1.apply(lambda row: row.Population + row.Population_change*5, axis = 1)
df1.head()
```
